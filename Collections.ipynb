## Deques || Double Ended queues || decks

```
''' Deque works the same way like a list but deque has memory efficient appends and are faster when we need to insert items at the beginning ''' 
from collections import deque as dq
```

```
''' Creates a deque '''
a=dq('abc')
a
```

```
''' Adds a value to the right '''
a.append('d')
a
```

```
 ''' Adds a value to the left '''
 a.appendleft('0')
 a
```

```
''' For adding multiple item to the left of the deque '''
a.extend('efg')
a
```

```
''' For adding multiple item to the right of the deque '''
a.extendleft('123')
a
```

```
''' In order to remove an item from the right of the deque '''
a.pop()
a
```

```
''' In order to remove an item from the left of the deque '''
a.popleft()
a
```

```
''' In order to move the deque items n-step ahead or to the right '''
n=2
a.rotate(n)
a
```

```
''' In order to move the deque items n-step behind or to the left '''
n=-2
a.rotate(n)
a
```

```
''' The method is used in order to get only the selected item in the list '''
import itertools
list (itertools.islice(a,3,9))
```

```
''' Maxlen parameter restricts the size of the deque and makes it suitable for circular buffer data structure '''
b=dq([],maxlen=3)
for i in range (6):
  b.append(i)
  print(b)
```

# Chain Map

```
''' Chain Map are used to create an ordered list of dictionaries and also used to link multiple number of dictionaries into one object '''
from collections import ChainMap
dictionary1={'a':1,'b':2,'c':3,'d':4}
dictionary2={'e':5,'f':6,'g':7,'h':8}
c=ChainMap(dictionary1,dictionary2)
c
```

```
c.maps
```

```
c.values
```

```
''' For getting only a particular value '''
c['b']
```

```
''' Unlike dictionary where we can retain only previous set values, adding a child context in the overrides the same key but it does not remove from the Data Structure and we can actually view the history of changes and if we wish we can add the previous chages as well '''
d={'sky':'sky','sun':'orange','tree':'green','water':'transparent'}
d
```

```
'''  Creates a chain map with previous configurations '''
e=ChainMap(d)
```

```
e.values()
```

```
''' For creating a new child which override the parent '''
e1=e.new_child({'sky':'blue'})
#For returing the overriden theme 
e1['sky']
```

```
e1.pop('sky')
```

```
e1['sky']
```

```
e1.maps
```

```
e1.parents
```

# Counter Objects

```
''' Counter is a subclass of dictionary where each dictionary key is a hashable object and the associated value is an integer count of that object '''
from collections import Counter
Counter('AlekhayaDatta')
```

```
f1=Counter({'a':1,'b':1,'c':3})
f1
```

```
f2=Counter(a=1,c=2,b=1,d=3)
f2
```

```
# Creating the empty counter object
ct=Counter()
```

```
ct.update('Management Development Institute of Singapore')
ct
```

```
ct.update({'a':7})
ct
```

```
for item in ct:
  print('%s:%d'%(item,ct[item]))
```

```
#If an item is missing in the counter, then instead of returning a error it will return the zero count
ct['x']
```

```
ct2=Counter('Counter')
ct2
```

```
ct2.update({'C':-2})
```

```
sorted(ct2.elements())
```

```
ct3=Counter('HELLO WORLD')
ct3
```

```
#It takes a positive integer argument that determones the number of most common element to return
ct3.most_common()
```

```
#Instead of adding values, it subtracts the value
ct3.subtract({'E':2})
ct3
```

# Ordered Dictionaries

```
from collections import OrderedDict
a1=OrderedDict()
a1['one']=1
a1['two']=2
a2=OrderedDict()
a2['two']=2
a2['one']=1
a1==a2
```

```
c=[('three',3),('four',4),('five',5)]
a1.update(c)
a1
```

```
for k,v in a1.items():
  print(k,v)
```

```
c2=OrderedDict(sorted(a1.items(), key=lambda t: (4*t[1]-t[1]**2)))
c2
```

# defaultdict

```
#Unlike normal dictionary when the value is missing, instead of throwing an error- this method will override the missing key and creates a new instance variable
from collections import defaultdict
f1=defaultdict(int)
words=str.split('red green orange green red yellow')
for word in words:
  f1[word]+=1
f1
```

# namedtuple

```
'''namedtuple instances are just as memory efficient as regular tuples because they do not have per-instance dictionaries. 
Each kind of namedtuple is represented by its own class, created by using the namedtuple() factory function. 
The arguments are the name of the new class and a string containing the names of the elements.
from collections import namedtuple'''
Point = namedtuple('Point', 'x y')
pt1 = Point(1.0, 5.0)
pt2 = Point(2.5, 1.5)

from math import sqrt
line_length = sqrt((pt1.x-pt2.x)**2 + (pt1.y-pt2.y)**2)
print(line_length)
```
